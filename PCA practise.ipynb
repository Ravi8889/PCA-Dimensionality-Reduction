```python
df1=df.iloc[:,1:]
df1
```

```python
# Noramalization of data
def norm_data(i):
    x = (i - i.min())/(i.max()-i.min())
    return x
```

```python
normal_df =norm_data(df1)
```

```python
normal_df
```

```python
pca= PCA()
pca_values=pca.fit_transform(normal_df)
pca_values
```

```python
pca=  PCA(n_components= 5)
pca_val =pca.fit_transform(normal_df)
pca_val
```

```python
# Each pca  feature has how much information while creating the pca 
pca_ratio = pca.explained_variance_ratio_
pca_ratio
```

```python
# Each pca  feature has how much information while creating the pca 
pca_cummlative_ratio = np.cumsum(np.round(pca_ratio,decimals= 4)*100)
pca_cummlative_ratio
```

```python
# plot the pca values
plt.plot(pca_cummlative_ratio,color="green")
```
